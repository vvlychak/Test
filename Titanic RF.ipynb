```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic = pd.read_csv('C:/Users/vladi/Desktop/Программирование/Datasets/Titanic/train.csv')
titanic.head()
```

```python
X = titanic.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'], axis=1)
y = titanic.Survived
X
```

```python
X = pd.get_dummies(X)
X
```

```python
X.isnull().sum()
```

```python
X = X.fillna({'Age': X.Age.median()})
X.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
clf = RandomForestClassifier()
```

```python
parametrs = {'n_estimators': [10, 50, 100], 'criterion': ['gini', 'entropy'], 'max_depth': range(1, 30)}
```

```python
randomized_search_cv = RandomizedSearchCV(clf, parametrs, cv=5)
```

```python
randomized_search_cv.fit(X_train, y_train)
```

```python
randomized_search_cv.best_params_
```

```python
best_clf = randomized_search_cv.best_estimator_
best_clf
```

```python
best_clf.fit(X_train, y_train)
```

```python
best_clf.score(X_test, y_test)
```
